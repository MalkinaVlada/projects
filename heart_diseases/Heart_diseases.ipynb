# Greedy algorithm full pass 

```python
import numpy as np
import itertools
import statsmodels.api as sm

def exhaustive_feature_selection(X, y, criterion='r_squared'):
    num_features = X.shape[1]
    best_score = -np.inf if criterion == 'r_squared' else np.inf
    best_subset = None
    #feature_valuations = []
    
    for k in range(1, num_features + 1):
        subsets = itertools.combinations(range(num_features), k)
        
        for subset in subsets:
            model = sm.OLS(y, sm.add_constant(X[:, subset])).fit()
            
            if criterion == 'r_squared':
                score = model.rsquared
            elif criterion == 'mean_squared_error':
                score = -model.mse_model
            
            if (criterion == 'r_squared' and score > best_score) or (criterion == 'mean_squared_error' and score < best_score):
                best_score = score
                best_subset = subset
            #feature_valuations.append((len(subset), score))
    return list(best_subset)

# Usage example
best_subset = exhaustive_feature_selection(x_train, y_train, criterion='r_squared')
print("Best subset of features:", best_subset)
#print(feature_valuations)
```

# forward selection

```python
import numpy as np
import statsmodels.api as sm

def forward_feature_selection(X, y, p_value_threshold=0.05):
    num_features = X.shape[1]
    selected_features = []
    current_best_feature = None
    
    while True:
        remaining_features = list(set(range(num_features)) - set(selected_features))
        best_p_value = np.inf
        best_feature = None
        
        for feature in remaining_features:
            model = sm.OLS(y, sm.add_constant(X[:, selected_features + [feature]])).fit()
            p_value = model.pvalues[-1]  # Get p-value of the added feature
            
            if p_value < best_p_value:
                best_p_value = p_value
                best_feature = feature
        
        if best_feature is not None and best_p_value < p_value_threshold:
            selected_features.append(best_feature)
            current_best_feature = best_feature
        else:
            break
        print(selected_features)
    
    return list(selected_features)

# Usage example
selected_features = forward_feature_selection(x_train, y_train, p_value_threshold=0.05)
print("Selected features:", selected_features)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
def results(y_test, y_pred, model):
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.2f}")

    # Calculate F1 score
    f1 = f1_score(y_test, y_pred)
    print(f"F1 Score: {f1:.2f}")

    # Calculate precision
    precision = precision_score(y_test, y_pred)
    print(f"Precision: {precision:.2f}")

    # Calculate recall
    recall = recall_score(y_test, y_pred)
    print(f"Recall: {recall:.2f}")
    
    # Get feature importances
    importances = model.feature_importances_

    print()
    print('Feature importances')
    # Print feature importance scores
    for feature, importance in zip(col, importances):
        print(f"{feature}: {importance:.4f}")
```
